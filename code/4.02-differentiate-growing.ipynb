## Formation of the aggregate

### Initialization

```julia
parameters = define_par();              

dt = 0.002;                             
save_each = round(Int64, 0.25 / dt);    
n_cells = 60;                          
```

```julia
Random.seed!(2345)                              
com = initialize_growth(parameters; dt = dt);   
```

### Aggregation

```julia
grow_size!(com, save_each, n_cells)
m0 = length(com);
```

```julia
println(com.N)
println(formed_correctly(com))          
plot_aggregate(com, color_map, 1, m0)   
```

```julia
growncom = deepcopy(com);   setfield!(com,:dt, 0.001)            
```

## Differentiation

```julia
com = deepcopy(growncom)
Random.seed!(2345);
```

```julia
# setfield!(com,:dt, 0.001);
com.tau_div = 8;
```

```julia
initialize_diff!(com, g_on = true)                   
m1 = length(com);
```

```julia
differentiate_growing!(com, save_each, 30, prot=100)
m2 = length(com);
```

```julia
println(com.N)
println(formed_correctly(com))   
plot_aggregate(com, color_map, m1, m2)
```

```julia
evolvedcom = deepcopy(com);
```

## Dimensionalization

```julia
com = deepcopy(evolvedcom);
```

```julia
dimensionalize_com!(com);
```

```julia
plot_aggregate(com, color_map, m1, m2; showtime=true, shownumbers=false)
```
